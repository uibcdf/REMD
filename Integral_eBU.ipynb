```python
import numpy as np
import math
```

```python
Lmin=-100
Lmax=100
n_bins=np.arange(-100,100,0.01)
x_lista=[Lmin+ii*0.01 for ii in n_bins]
```

```python
def double_well(x,Eo,C,M):
    return Eo*(C*x**4-x**2)+M*x

Eo=6.0
C=0.4
M=-0.25
```

```python
Area = 0 
for kk in x_lista:
    Area=Area+(math.exp(-5.21e-21)*(double_well(kk,Eo,C,M))*0.01)
    print(Area)
```
